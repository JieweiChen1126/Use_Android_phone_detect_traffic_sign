```python
# ================================
# 步驟 3：建立 dataset.yaml
# ================================
import yaml

# 修改這裡：你的類別數與名稱
class_names = ['Crossing', 'speed_30','speed_40','speed_50','speed_60','speed_70','30','40','50','60','70','red_light','green_light','yellow_light',
         'forbidden','take_photo','two_left']  # ← 替換成你的類別名稱
num_classes = len(class_names)

dataset_yaml = {
    'path': '/content/drive/My Drive/Colab Notebooks/yolo_dataset',
    'train': 'images/train',
    'val': 'images/val',
    'nc': num_classes,
    'names': class_names
}

with open('/content/drive/My Drive/Colab Notebooks/yolo_dataset/dataset.yaml', 'w') as f:
    yaml.dump(dataset_yaml, f)

print("✅ 已建立 dataset.yaml")

```

```python
from ultralytics import YOLO

# 修改這裡：你的類別名稱要和 dataset.yaml 裡的一樣
class_names = ['Crossing', 'speed_30','speed_40','speed_50','speed_60','speed_70','30','40','50','60','70','red_light','green_light','yellow_light',
         'forbidden','take_photo','two_left']  # ← 請改成你自己的類別名稱！

# 載入模型
model = YOLO('/content/drive/My Drive/Colab Notebooks/yolo_dataset/yolo_results/my_yolo_model_16/weights/best.pt')

# 測試圖片路徑（你可以替換成其他圖片）
image_path = '/content/drive/My Drive/Colab Notebooks/yolo_dataset/test'

# 執行預測
results = model.predict(source=image_path, save=True, conf=0.1)


# 逐一印出預測結果
for r in results:
    print(f"\n🔍 預測結果 for {image_path}:")
    for i, box in enumerate(r.boxes):
        cls_id = int(box.cls[0])
        cls_name = class_names[cls_id] if cls_id < len(class_names) else f"Unknown({cls_id})"
        conf = float(box.conf[0])
        xyxy = [round(x, 2) for x in box.xyxy[0].tolist()]
        print(f"  [{i+1}] 類別：{cls_name}，信心值：{conf:.2f}，邊界框：{xyxy}")
```

```python
# 載入你訓練好的模型
from ultralytics import YOLO

# 確保你已經載入了你的 best.pt 模型
model = YOLO('/content/drive/My Drive/Colab Notebooks/yolo_dataset/yolo_results/my_yolo_model_16/weights/best.pt')

# 匯出為 TFLite 格式
# imgsz=640 是指定匯出後的模型輸入尺寸，這很重要！
# int8=True 是進行量化，可以讓模型更小、更快，但可能稍微犧牲精度。建議先從非量化的 (float32) 開始。
model.export(format='tflite', imgsz=640)

# 如果想進行 INT8 量化，你需要一個代表性的資料集
# !yolo export model=... format=tflite imgsz=640 int8=True data=dataset.yaml
```
